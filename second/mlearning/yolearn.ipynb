```python
result = [fname[:-4] for fname in os.listdir("data/images") if fname.endswith(".jpg")]
```

```python
def replace(ls, path):
    for name in ls:
        os.rename(f"data/images/{name}.jpg", f"{path}/images/{name}.jpg")
        os.rename(f"data/labels/{name}.txt", f"{path}/labels/{name}.txt")
```
